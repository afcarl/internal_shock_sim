## Initial Parameters

```python
TOTAL_ENERGY = 2*1.E52/(4* pi)
dT = .002
t0 = 0.
Rmin = 1E10
tw=10


NUM_SHELLS = int(tw/dT)

print NUM_SHELLS

dE = float(TOTAL_ENERGY)/float(NUM_SHELLS)
```
